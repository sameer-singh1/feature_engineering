##### Target guided ordinal- It is the process of replacing a categorical value with the mean of the target variable.

```python
## traget guided ordinal

df.groupby('Airline')['Price'].mean().sort_values()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

#### One-hot encoding is essentially the representation of categorical variables as binary vectors. These categorical values are first mapped to integer values. Each integer value is then represented as a binary vector that is all 0s (except the index of the integer which is marked as 1).

```python
ohe=OneHotEncoder()
```

```python
ohe.fit_transform(df[['Airline']]).toarray()
```

```python
## Replacing target guided ordinal encoding
def replace_airline_with_mean(df):
    mean_prices = df.groupby('Airline')['Price'].mean().sort_values()
    df['Airline'] = df['Airline'].apply(lambda x: mean_prices[x])
    return df

df = replace_airline_with_mean(df)
df.head()
```
